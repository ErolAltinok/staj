NESNE EMİN OLMA OLASILIĞI %20'den küçükse yazdırmamak için kodlar.

```python
img = cv2.imread('C:/Users/reybe/Desktop/staj/YOLOv8 foto/2.jpg')
rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
results = model(rgb_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX #Resimde göstereceğimiz yazıların fontunu böyle ayarlıyoruz.

for i in range(len(results[0].boxes)): #resimde ne kadar nesne varsa for döngüsü o kadar dönecek.
    x1,y1,x2,y2=results[0].boxes.xyxy[i] # her bir nesne için konumlarını alıyoruz.
    score=results[0].boxes.conf[i] #score bilgilerini alıyoruz.
    label=results[0].boxes.cls[i] #label bilgilerini alıyoruz.
    x1,y1,x2,y2,score,label=int(x1),int(y1),int(x2),int(y2),float(score),int(label)
    #bunları uygun formata çeviriyoruz. normalde tensor formatındalar.
    #score 0 ile 1 arası olabileceği için onu float yaptık.
    name=labels[label] #label integer değerinde olduğu için(yukardaki results kısmındaki names'de gördük.)
    #labelsi bu şekilde string hale getirdik.
    if score<0.2: #böyle bir if ekledik çünkü nesnenin emin olma durumu yani score'u %20'den küçük olursa onu resimde göstermeyecek.
        continue

    cv2.rectangle(img,(x1,y1),(x2,y2),(255,0,0),2) # bu komut ile her nesneye bir sınırlayıcı kutu çizdiriyoruz.
    text= name+' '+str(format(score, '.2f')) # nesnenin adını ve score bilgisini yazdırıyoruz.
    cv2.putText(img, text,(x1, y1-10), font, 1.2, (255,0,255), 2) #putText ile de ekranda gösteriyoruz.
   #(255,0,0) => renkler. 2'ler kalınlık. 

cv2.imshow("kamera",img) # ekranda gösteriyoruz.

cv2.waitKey()#klavyeden bir tuşa basılmasını bekler.
cv2.destroyAllWindows()#klavyeden bir tuşa basıncada ekranın kapanması için.
```

RESİMDE SADECE BELİRLEDĞİMİZ SINIFIN GÖZÜKMESİ

```python
img = cv2.imread('C:/Users/reybe/Desktop/staj/YOLOv8 foto/3.jpg')
rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
results = model(rgb_img)

font = cv2.FONT_HERSHEY_SIMPLEX

for i in range(len(results[0].boxes)): 
    x1,y1,x2,y2=results[0].boxes.xyxy[i] 
    score=results[0].boxes.conf[i] 
    label=results[0].boxes.cls[i] 
    x1,y1,x2,y2,score,label=int(x1),int(y1),int(x2),int(y2),float(score),int(label)
    name=labels[label]
    if score<0.5 or name!="person": #Sadece insanı gösterecek.
        continue

    cv2.rectangle(img,(x1,y1),(x2,y2),(255,0,0),2)
    text= name+' '+str(format(score, '.2f'))
    cv2.putText(img, text,(x1, y1-10), font, 1.2, (255,0,255), 2)
    
cv2.imshow("kamera",img)

cv2.waitKey()
cv2.destroyAllWindows()
```

Videolarda nesne tespiti

```python
kamera = cv2.VideoCapture('C:/Users/reybe/Desktop/staj/YOLOv8 foto/trafik video.mp4')
#VideoCapture ile videoyu aldık.
font = cv2.FONT_HERSHEY_SIMPLEX

while True:

    ret,kare=kamera.read()
    #ret = kameradan görüntü gelip gelmediğini söylüyor.
    #kare = her bir frame'de gelen görüntü.
    
    imgs=cv2.cvtColor(kare,cv2.COLOR_BGR2RGB)
    results = model(imgs,verbose=False)
    
    for i in range(len(results[0].boxes)):
        x1,y1,x2,y2=results[0].boxes.xyxy[i]
        score=results[0].boxes.conf[i]
        label=results[0].boxes.cls[i]
        x1,y1,x2,y2,score,label=int(x1),int(y1),int(x2),int(y2),float(score),int(label)
        name=labels[label]
        if score<0.5:
            continue

        cv2.rectangle(kare,(x1,y1),(x2,y2),(255,0,0),2)
        text= name+' '+str(format(score, '.2f'))
        cv2.putText(kare, text,(x1, y1-10), font, 1.2, (255,0,255), 2)
    
    cv2.imshow("kamera",kare)
    if cv2.waitKey(1) & 0xFF == ord('q'): #her bir saniyede bekle ve q tuşuna basarsak çık.
        break

kamera.release()
cv2.destroyAllWindows()

#ÖNEMLİ !!! videonun daha akıcı olması için başta YOLOv8 modelini belirlerken
#yolov8m yerine yolov8n yaparsak video daha akıcı olur.
```

Videoda nesne takibi.

```python
from ultralytics import YOLO
model = YOLO("yolov8l.pt")

import os
os.environ["KMP_DUPLICATE_LIB_OK"]="TRUE" #Ram'in dolmaması için eklenebilir kod.
```

```python
kamera = cv2.VideoCapture('C:/Users/reybe/Desktop/staj/YOLOv8 foto/trafik video.mp4')
font = cv2.FONT_HERSHEY_SIMPLEX

while True:

    ret,kare=kamera.read()
    
    imgs=cv2.cvtColor(kare,cv2.COLOR_BGR2RGB)
    results = model.track(imgs,persist=True,verbose=False) #.track ekliyoruz çünkü takip modunda olduğumuzu belirtiyoruz.
    #persist=true yapıyoruz çünkü her frame değiştiğinde en baştan id atamasın diye. 
    
    for i in range(len(results[0].boxes)):
        x1,y1,x2,y2=results[0].boxes.xyxy[i]
        score=results[0].boxes.conf[i]
        label=results[0].boxes.cls[i]
        ids=results[0].boxes.id[i] # burda id'leri alıyoruz.
        x1,y1,x2,y2,score,label,ids=int(x1),int(y1),int(x2),int(y2),float(score),int(label),int(ids) #int(ids) eklemeyi unutma.
        name=labels[label]
        if score<0.2:
            continue

        cv2.rectangle(kare,(x1,y1),(x2,y2),(255,0,0),2)
        text= name+' '+str(ids)+' '+str(format(score, '.2f'))
        cv2.putText(kare, text,(x1, y1-10), font, 1.2, (102,0,153), 2)
    
    cv2.imshow("kamera",kare)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

kamera.release()
cv2.destroyAllWindows()
```
